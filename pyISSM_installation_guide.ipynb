2. Start a JupyterLab session with the correct settings
3. Open a Terminal inside JupyterLab
4. Navigate to your project folder

6. Install pyISSM
7. Install the ACCESS Cryosphere Data Pool
8. Verify that both installed correctly

# PART 1 — The ARE Portal

---

## Step 1: Login to ARE

ARE stands for **Australian Research Environment**. It's a website that gives you a friendly graphical interface to work on Gadi, without needing to be a command-line expert. Think of it as a web browser-based window into the supercomputer.

**To log in:**

1. Open your web browser and go to: **[https://are.nci.org.au](https://are.nci.org.au)**
2. Click **Log in** and enter your **NCI username and password**.

> **Don't have an NCI account yet?**  
> You need to be registered with NCI and added to the project `su58` before you can log in. You should also apply to the projects `xp65`, `vk83`, `av17`, and `au88`.

---

## Step 2: Start a JupyterLab Session

Once you're logged in to ARE, you'll see a dashboard from which you can launch a **JupyterLab** session. 

**To start a session:**

1. On the ARE dashboard, click on **JupyterLab**.
2. A form will appear asking you to configure your session. This is important; it tells Gadi what software to load and which data storage folders you need access to.
3. Fill in the form with **exactly** the values in the table below.

---

### Required Session Settings

| Field | What to enter |
|---|---|
| **Project** | `su58` |
| **Storage** | `gdata/xp65+gdata/au88+gdata/vk83+scratch/au88` |
| **Module directories** | `/g/data/vk83/modules /g/data/xp65/public/modules` |
| **Modules** | `access-issm/2025.11.0 conda/analysis3` |

> **What do these settings mean?**  
> - **Project**: The research project you belong to on Gadi. This is how computing time gets billed. YOU MUST MAKE SURE YOU BELONG TO THE PROJECT YOU SELECT.
> - **Storage**: The disk locations on Gadi that your session needs to read/write. `gdata` is long-term project storage; `scratch` is fast temporary workspace. The `+` just means "and also".  
> - **Module directories**: Gadi uses a system called "modules" to manage software installations. These are the folder paths where this project's modules are kept.  
> - **Modules**: The specific software to load -- `access-issm` -- provides the ISSM model itself and `conda/analysis3` gives you a Python environment with many science packages pre-installed.

4. Once the form is filled in, click **Launch**
5. You'll see your session listed as **Queued** or **Starting**. This can take a minute or two. Once it shows **Running**, click **Open JupyterLab**.

> **If it takes more than 5 minutes to start**, that's normal. Gadi is a shared supercomputer and sometimes you wait briefly for resources. Just be patient!

# PART 2 — Opening a Terminal

---

## Step 3: Open a Terminal in JupyterLab

You're now inside JupyterLab on Gadi. The installation steps for pyISSM and the ACCESS Cryosphere Data Pool need to be done through a **Terminal**, a text-based window where you type commands directly to the computer.

**To open a Terminal:**

1. In JupyterLab, look for the **Launcher** tab (it shows a grid of icons for things you can open). If you don't see it, go to **File → New Launcher** in the top menu.
2. Scroll down to the section labelled **Other** and click the **Terminal** icon.
3. A new tab will open that should be a dark window with a blinking cursor. This is your Terminal!

> **What is a Terminal?**
> A terminal (also called a "command line" or "shell") lets you control the computer by typing text commands instead of clicking buttons. It might look intimidating at first, but you only need to type the specific commands shown in this guide. Take it one step at a time.

> **Important:** **Part 3 of this guide happens in the Terminal window**, not here in this notebook. Once you've completed the Terminal steps, come back to this notebook for Part 4 to verify everything worked.

---

# PART 3 — Installing pyISSM via the Terminal

The steps below all happen in the **Terminal tab** you just opened. Type (or copy-paste) each command carefully and press **Enter** to run it.

---

## Step 4: Navigate to Your Project Folder

On Gadi, your files are stored in a specific folder location. The `cd` command stands for **"change directory"**. It moves you from one folder to another in the terminal, just like clicking through folders in a file explorer on your laptop.

In the Terminal, type the command below **replacing the placeholders with your actual details:**

```bash
cd /g/data/<projectname>/<userid>/
```

**Replace:**
- `<projectname>` → your NCI project code (most likely `su58`)
- `<userid>` → your NCI username, usually a combination of your initials and some numbers (e.g., `abc123`)

**Worked example** (don't copy this exactly; use your own username!):
```bash
cd /g/data/su58/abc123/
```

After pressing Enter, the terminal prompt will update to show your new location. You might not see any other message and that's completely normal! In the terminal, **silence usually means success**.

> **Tip:** At any point, you can type `pwd` and press Enter to see exactly which folder you're currently in (`pwd` = "print working directory"). You can also type `ls` to **l**i**s**t the files and folders in your current location.


If you get an error saying that the directory does not exist, you will need to create a directory. In that case, you should type the commands below **replacing the placeholders with your actual details:**
```bash
cd /g/data/<projectname>/
mkdir <userid>
cd <userid>
```

---

## Step 5: Download pyISSM from GitHub

Now we'll download the pyISSM source code from the internet. We use a command called **`git clone`**, think of it like downloading a folder from the web, but in a way that preserves the full history of the code.

In the Terminal, type:

```bash
git clone https://github.com/ACCESS-NRI/pyISSM.git
```

Press **Enter**. You'll see output like:

```
Cloning into 'pyISSM'...
remote: Enumerating objects: ...
Receiving objects: 100% ...
```

This means the download was successful. A new folder called `pyISSM` has been created in your current directory.

> **What is GitHub?**  
> GitHub is a popular website where software developers share and store code. `git clone` downloads a full copy of that code to wherever you're working, in this case, your folder on Gadi.

---

## Step 6: Move into the pyISSM Folder

Navigate into the folder that was just downloaded:

```bash
cd pyISSM
```

Press **Enter**. Again, no confirmation message is normal, you're now inside the `pyISSM` folder.

> Try typing `ls` and pressing Enter to see what's inside this folder. You should see files like `setup.py` or `pyproject.toml`, these are the files that tell pip how to install the package.

---

## Step 7: Install pyISSM

Now we install the package using **`pip`**, Python's package installer. The `.` at the end of the command means "install the package from the **current folder**" (i.e., the `pyISSM` folder you just moved into).

```bash
pip install .
```

Press **Enter**. This will take a little while and you'll see a lot of text scrolling past as pip downloads and installs pyISSM and everything it depends on. This is normal!

Wait until you see something like:
```
Successfully installed pyissm-...
```
That means pyISSM has been installed.

> **If you see red text:**  
> Don't panic. Not all red text means something is broken. Look specifically for the word `ERROR`. If you see it, double-check that you're inside the `pyISSM` folder (run `pwd` to confirm) and try again. If errors persist, take a screenshot and ask your supervisor or demonstrator.

> **Common mistake:** If you get an error like `No file/folder found`, it usually means you forgot to run `cd pyISSM` in Step 6 before running `pip install .`. Go back and check!

---

## Step 8: Install the ACCESS Cryosphere Data Pool

The **ACCESS Cryosphere Data Pool** (`access-cryosphere-data-pool`) is a companion package that gives you programmatic access to curated ice and climate datasets already stored on Gadi. You'll need it for your modelling work.

This one is essentially the same as the pyISSM installation. Again, opening a terminal in a jupyter notebook on ARE (you can use the same one you used for pyISSM before), run the following:

```bash
git clone https://github.com/ACCESS-NRI/access-cryosphere-data-pool.git

cd access-cryosphere-data-pool

pip install .
```

Press **Enter** and wait for it to finish. Look for `Successfully installed access-cryosphere-data-pool-...` at the end.

> **What is this package for?**  
> Instead of manually hunting through Gadi's file system for the right dataset files, `access-cryosphere-data-pool` lets you browse and load ice/snow datasets directly in Python with just a few lines of code.

> **Note:** This package is still under active development. Some documentation may be incomplete; that's okay, the core functionality you need is stable.

---

**The Terminal steps are now complete!**

Head back to this notebook and continue with Part 4 below to verify your installation.

# PART 4 — Verifying Your Installation

Now we're back in the notebook. You can close the Terminal tab if you like.

We'll run some quick checks to make sure both packages installed correctly. **Run each code cell below by clicking on it and pressing Shift + Enter.**

---

## Step 9: Check that pyISSM installed correctly

If pyISSM is installed correctly, the cell below will print the version number. If you get a red `ModuleNotFoundError`, go back and repeat Steps 6–7.

````python
# 'import' loads the package into Python so we can use it.
# If this cell runs without any error, pyISSM is installed correctly!

import pyissm

print("pyISSM imported successfully!")
print("pyISSM version:", pyissm.__version__)
````

**Expected output:**
```
pyISSM imported successfully!
pyISSM version: 0.x.x
```
(The exact version number will vary.)

> **If you get `ModuleNotFoundError: No module named 'pyissm'`:**  
> This means Python can't find pyISSM. The most common cause is that `pip install .` in Step 7 didn't complete successfully, or it was run from the wrong folder. Go back to the Terminal, make sure you're inside the `pyISSM` folder (`cd /g/data/su58/<userid>/pyISSM`), and try `pip install .` again.

---

## Step 10: Check that the Cryosphere Data Pool installed correctly

Run the cell below. If it prints a version number without errors, the data pool package is ready to use.

````python
# The access-cryosphere-data-pool package is imported under the name 'datapool'.
# Note: the *install* name and the *import* name are different — this is common in Python!

import ccdtools

print("access-cryosphere-data-pool imported successfully!")
print("datapool version:", ccdtools.__version__)
````

**Expected output:**
```
access-cryosphere-data-pool imported successfully!
datapool version: 0.x.x
```

> **If you get `ModuleNotFoundError: No module named 'datapool'`:**  
> Go back to the Terminal and re-run `pip install access-cryosphere-data-pool` from Step 8. Make sure that you're a member of the project `av17` -- without that python won't be able to find any of the relevant datasets.

---

## Step 11: Full installation summary check

Run the cell below for a combined summary of both installations, including where the packages are stored on Gadi. This is useful information if you ever need to troubleshoot.

````python
# A combined summary check of both packages.
# __file__ tells us where on the filesystem the package is installed.

import pyissm
import ccdtools

print("=" * 55)
print("         Installation Summary")
print("=" * 55)
print(f"  pyISSM version   : {pyissm.__version__}")
print(f"  pyISSM location  : {pyissm.__file__}")
print()
print(f"  datapool version : {ccdtools.__version__}")
print(f"  datapool location: {ccdtools.__file__}")
print("=" * 55)
print()
print("Everything is installed and ready to go!")
````

## Troubleshooting Summary

| Problem | What to check |
|---|---|
| Can't log in to ARE | Make sure you've been added to project `su58` by your supervisor |
